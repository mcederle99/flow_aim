```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ep_steps = np.load('ep_steps_big_random_simple_1_2.npy')
returns = np.load('returns_big_random_simple_1_2.npy')
num_eps = np.arange(len(ep_steps))
```

```python
ema_ret = pd.Series(returns).ewm(span=30, adjust=False).mean()

plt.plot(num_eps*5000, ema_ret);
plt.plot(num_eps*5000, returns);

plt.show()
```

```python
ema_steps = pd.Series(ep_steps).ewm(span=30, adjust=False).mean()

plt.plot(num_eps*5000, ema_steps);
plt.plot(num_eps*5000, ep_steps);

plt.show()
```
