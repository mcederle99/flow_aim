```python
import numpy as np
```

```python
ep_steps = np.load('ep_steps_random_simple_2.npy')
returns = np.load('returns_random_simple_2.npy')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
num_eps = np.arange(len(ep_steps))
```

```python
ema_steps = pd.Series(ep_steps).ewm(span=10, adjust=False).mean()

plt.plot(num_eps, ema_steps);

plt.show()
```

```python
ema_ret = pd.Series(returns).ewm(span=1, adjust=False).mean()

plt.plot(num_eps, returns);

plt.show()
```
