```python
import numpy as np
```

```python
ep_steps = np.load('ep_steps.npy')
returns = np.load('returns.npy')
```

```python
ep_steps_clip = np.load('ep_steps_clip.npy')
returns_clip = np.load('returns_clip.npy')
```

```python
returns_per_veh_clip = np.load('returns_per_veh_clip.npy')
returns_per_veh = np.load('returns_per_veh.npy')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
num_eps = np.arange(len(ep_steps))
num_eps_clip = np.arange(len(ep_steps_clip))
```

```python
ema_steps = pd.Series(ep_steps).ewm(span=30, adjust=False).mean()
ema_steps_clip = pd.Series(ep_steps_clip).ewm(span=30, adjust=False).mean()

plt.plot(num_eps, ema_steps);
plt.plot(num_eps_clip, ema_steps_clip);

plt.show()
```

```python
ema_ret = pd.Series(returns).ewm(span=30, adjust=False).mean()
ema_ret_clip = pd.Series(returns_clip).ewm(span=30, adjust=False).mean()

plt.plot(num_eps, ema_ret);
plt.plot(num_eps_clip, ema_ret_clip);

plt.show()
```

```python
ema_retxveh = pd.Series(returns_per_veh).ewm(span=30, adjust=False).mean()
ema_retxveh_clip = pd.Series(returns_per_veh_clip).ewm(span=30, adjust=False).mean()

plt.plot(num_eps, ema_retxveh);
plt.plot(num_eps_clip, ema_retxveh_clip);

plt.show()
```
