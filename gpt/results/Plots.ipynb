```python
import numpy as np
```

```python
ep_steps_curr = np.load('ep_steps_curr.npy')
returns_curr = np.load('returns_curr.npy')
```

```python
ep_steps2 = np.load('ep_steps2.npy')
returns2 = np.load('returns2.npy')
```

```python
ep_steps_random = np.load('ep_steps_random.npy')
returns_random = np.load('returns_random.npy')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
num_eps_curr = np.arange(len(ep_steps_curr))
#num_eps2 = np.arange(len(ep_steps2))
num_eps_random = np.arange(len(ep_steps_random))
```

```python
ema_steps_curr = pd.Series(ep_steps_curr).ewm(span=30, adjust=False).mean()

plt.plot(num_eps_curr, ema_steps_curr);
```

```python
ema_steps_2 = pd.Series(ep_steps2).ewm(span=30, adjust=False).mean()

plt.plot(num_eps2, ema_steps_2);
```

```python
ema_steps_random = pd.Series(ep_steps_random).ewm(span=30, adjust=False).mean()

plt.plot(num_eps_random, ema_steps_random);
```

```python
ema_ret_random = pd.Series(returns_random[:len(num_eps_curr)]).ewm(span=30, adjust=False).mean()
ema_ret_curr = pd.Series(returns_curr).ewm(span=30, adjust=False).mean()

plt.plot(num_eps_curr, ema_ret_curr)

plt.plot(num_eps_curr, ema_ret_random)

plt.show()
```

```python
ema_steps_random = pd.Series(ep_steps_random[:len(num_eps_curr)]).ewm(span=30, adjust=False).mean()
ema_steps_curr = pd.Series(ep_steps_curr).ewm(span=30, adjust=False).mean()

plt.plot(num_eps_curr, ema_steps_curr)

plt.plot(num_eps_curr, ema_steps_random)

plt.show()
```
