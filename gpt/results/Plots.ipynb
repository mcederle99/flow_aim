```python
import numpy as np
```

```python
ep_steps = np.load('ep_steps.npy')
returns = np.load('returns.npy')
```

```python
ep_steps2 = np.load('ep_steps2.npy')
returns2 = np.load('returns2.npy')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
num_eps = np.arange(len(ep_steps))
num_eps2 = np.arange(len(ep_steps2))
```

```python
ema_steps_2 = pd.Series(ep_steps2).ewm(span=30, adjust=False).mean()

plt.plot(num_eps2, ema_steps_2);
```

```python
ema_steps_1 = pd.Series(ep_steps).ewm(span=30, adjust=False).mean()

plt.plot(num_eps, ema_steps_1);
```

```python
ema_ret_1 = pd.Series(returns).ewm(span=30, adjust=False).mean()

plt.plot(num_eps, ema_ret_1);
```

```python
ema_ret_2 = pd.Series(returns2).ewm(span=30, adjust=False).mean()

plt.plot(num_eps2, ema_ret_2);
```
