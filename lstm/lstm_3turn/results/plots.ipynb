```python
import numpy as np
```

```python
ep_steps = np.load('batch/ep_steps_morecol.npy')
returns = np.load('batch/returns_morecol.npy')
returns_per_veh = np.load('batch/returns_per_veh_morecol.npy')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
num_eps = np.arange(len(ep_steps))
```

```python
ema_steps = pd.Series(ep_steps).ewm(span=30, adjust=False).mean()

plt.plot(num_eps, ema_steps);
#plt.plot(num_eps, ep_steps);

plt.show()
```

```python
ema_ret = pd.Series(returns).ewm(span=30, adjust=False).mean()

plt.plot(num_eps, ema_ret);
#plt.plot(num_eps, returns);

plt.show()
```

```python
ema_retxveh = pd.Series(returns_per_veh).ewm(span=30, adjust=False).mean()

plt.plot(num_eps, ema_retxveh);
#plt.plot(num_eps, returns_per_veh);

plt.show()
```
