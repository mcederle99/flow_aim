```python
import numpy as np
```

```python
ep_steps = np.load('ep_steps_fb.npy')
returns = np.load('returns_fb.npy')
returns_per_veh = np.load('returns_per_veh_fb.npy')
```

```python
ep_steps_gru = np.load('ep_steps_gru.npy')
returns_gru = np.load('returns_gru.npy')
returns_per_veh_gru = np.load('returns_per_veh_gru.npy')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
num_eps = np.arange(len(ep_steps))
#num_eps_gru = np.arange(len(ep_steps_gru))
```

```python
ema_steps = pd.Series(ep_steps).ewm(span=30, adjust=False).mean()
#ema_steps_gru = pd.Series(ep_steps_gru).ewm(span=30, adjust=False).mean()

plt.plot(num_eps, ema_steps);
#plt.plot(num_eps_gru, ema_steps_gru);

plt.show()
```

```python
ema_ret = pd.Series(returns).ewm(span=30, adjust=False).mean()
#ema_ret_gru = pd.Series(returns_gru).ewm(span=30, adjust=False).mean()

plt.plot(num_eps, ema_ret);
#plt.plot(num_eps_gru, ema_ret_gru);

plt.show()
```

```python
ema_retxveh = pd.Series(returns_per_veh).ewm(span=30, adjust=False).mean()
#ema_retxveh_gru = pd.Series(returns_per_veh_gru).ewm(span=30, adjust=False).mean()

plt.plot(num_eps, ema_retxveh);
#plt.plot(num_eps_gru, ema_retxveh_gru);

plt.show()
```
