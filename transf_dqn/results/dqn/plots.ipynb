```python
import numpy as np
```

```python
ep_steps = np.load('ep_steps.npy')
returns = np.load('returns.npy')
returns_per_veh = np.load('returns_per_veh.npy')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
num_eps = np.arange(len(ep_steps))
num_eps
```

```python
ema_steps = pd.Series(ep_steps).ewm(span=30, adjust=False).mean()

plt.plot(num_eps, ep_steps);

plt.show()
```

```python
ema_ret = pd.Series(returns).ewm(span=30, adjust=False).mean()

plt.plot(num_eps, ema_ret);

plt.show()
```

```python
ema_retxveh = pd.Series(returns_per_veh).ewm(span=30, adjust=False).mean()

plt.plot(num_eps, ema_retxveh);

plt.show()
```
